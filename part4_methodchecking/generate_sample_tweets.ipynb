```python
path = "../data/"

import pandas as pd


files = ["neo4j_nodos_16.csv","neo4j_nodos_20.csv","neo4j_nodos_24.csv"]
years = [int(nombre[-6:-4]) for nombre in files]

list_df =  [pd.read_csv(path+nombre).dropna() for nombre in files]
for i,df in enumerate(list_df):
    df["year"] = years[i]
df = pd.concat(list_df)
df
```

```python
df.contagio.value_counts()
```

```python
df2 = df.groupby('contagio', group_keys=False).sample(n=50)
```

```python
df2.contagio.value_counts()
```

```python
df2.tipo.value_counts()
```

```python
df2.to_csv(path+"muestra_contagio.csv",index=False)
```
